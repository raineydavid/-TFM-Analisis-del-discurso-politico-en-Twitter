```python
import pandas as pd
import re
from nltk.corpus import stopwords
from string import digits
import numpy as np
```

```python
a =pd.read_csv('train_tweet.csv')
```

```python
%%time
from gensim.models.keyedvectors import KeyedVectors
wordvectors_file_vec = 'fasttext-sbwc.3.6.e20.vec'
cantidad = 300000
wordvectors = KeyedVectors.load_word2vec_format(wordvectors_file_vec, limit=cantidad)
```

```python
def preparacióntexto(words):
    
    spanish_stopwords = stopwords.words('spanish')
    words = words.strip()
    words = words.lower()
    words = re.split(r'\W+', words)
    words_empty= [i for i in words if not i == '']
    sentence= [w for w in words_empty if not w in spanish_stopwords]        
    finally_sentence = [w for w in sentence if not w in digits]
    return finally_sentence
```

```python
def context_vector(words):
    tweet= pd.DataFrame()                               # crear dataframe vacio
    t1=[]                                               # crear lista vacia
    
    words= preparacióntexto(words)                      # Preparar el texto devuelve lista de palabras
                                                        
    for i in words:                                     # iterar sobre la lista de palabras 
        try:
            if not i == '':
                tweet_= wordvectors.get_vector(word=i)      # obtener el vector de cada palabra
                tweet_= pd.DataFrame(tweet_)                # crear un dataframe con el vector de cada palabra
                tweet_=tweet_.transpose()                   # transponer la columna en fila
                tweet= pd.concat([tweet,tweet_])            # añadirlo al dataframe vacio
            else:
                print(words)
        except:
            print('no esta en la lista de palabras' + i)
            pass
    
    for column in tweet:                                # iterar sobre las columnas del dataframe
        t1.append(tweet[column].mean())                 # añadir a la lista vacia la media de las columnas
        
    sentences= pd.DataFrame(t1)                         # crear un dataframe con la media de las columas
    sentences=sentences.transpose()                     # convertir la columna en fila
    
    return sentences                                    # devuelve un data frame de contexto 
```

```python
def context_data_frame(dt):
    context_data_frame=pd.DataFrame()
    for i in dt:
        sentences= context_vector(i)
        context_data_frame=pd.concat([context_data_frame,sentences])
    return context_data_frame
        
```
