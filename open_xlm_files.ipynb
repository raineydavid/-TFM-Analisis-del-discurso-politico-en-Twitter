```python
import pandas as pd

import urllib.request, urllib.parse, urllib.error
import xml.etree.ElementTree as ET
```

```python
url = 'http://www.sepln.org/workshops/tass/tass_data/dataset/intertass-ES-train-tagged.xml'

uh = urllib.request.urlopen(url)
data = uh.read()

tree = ET.fromstring(data)

lst = tree.findall('polarity')

counts = tree.findall('polarity')

for each in counts:
    print(each.text)
```

```python
tree
```

```python
tweet_id = []
content=[]
polarity=[]
for tweet in tree.iter('tweet'):
    tweet_id.append(tweet.find('tweetid').text)
    content.append(tweet.find('content').text)
for tweet in tree.iter('polarity'):
    polarity.append(tweet.find('value').text)

tweet_id=pd.DataFrame(tweet_id,columns=['tweet_id'])
content=pd.DataFrame(content,columns=['content'])
polarity=pd.DataFrame(polarity,columns=['polarity'])
train_tweet = pd.concat([tweet_id,content,polarity], axis= 1)
```

```python
train_tweet.to_csv('train_tweet.csv')
```
